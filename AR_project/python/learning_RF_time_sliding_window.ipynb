```python
#이 소스코드는 시간 기반 슬라이딩 윈도우 방식으로 Ramdom Forest를 학습하고 모델을 저장합니다.
#정리된 4일치 데이터를 학습하여 모델을 생성합니다.

import pandas as pd
#read time base slding window dataset
features = pd.read_csv('C:/Users/Shin/Documents/testbeddata/tssdata_16_19.csv')
features.head(5)
```

```python
import numpy as np
labels = np.array(features['state'])
features= features.drop(['state','timestamp'], axis = 1) #학습에 불필요한 state, timestamp 제거

feature_list = list(features.columns)
features = np.array(features)
```

Trainset 80%, Testset 20%

```python
from sklearn.model_selection import train_test_split
#trainset 80%, testset 20%
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
#Randmom Forest로 학습
rf = RandomForestClassifier(n_estimators = 200) #나무 200개
rf.fit(train_features, train_labels);
predictions = rf.predict(test_features)

from sklearn.metrics import confusion_matrix
confusion_matrix(test_labels, predictions)
```

```python
#confusion matrix and summary report
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
print('Accuracy Score :',accuracy_score(test_labels, predictions))
print('Report : ')
print(classification_report(test_labels, predictions))
```
