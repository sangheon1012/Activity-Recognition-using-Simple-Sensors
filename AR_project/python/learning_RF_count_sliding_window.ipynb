Read csv file, set feature and label

```python
import pandas as pd
#read count base slding window dataset
features = pd.read_csv('C:/Users/Shin/Documents/testbeddata/tscdata_16_19.csv')
features.head(5)
```

```python
import numpy as np
labels = np.array(features['state'])
features= features.drop(['state','timestamp'], axis = 1)

feature_list = list(features.columns)
features = np.array(features)
```

make Train set and Test set

```python
from sklearn.model_selection import train_test_split
#trainset 80%, testset 20%
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
#lets start learning
rf = RandomForestClassifier(n_estimators = 200)
rf.fit(train_features, train_labels);
predictions = rf.predict(test_features)

from sklearn.metrics import confusion_matrix
confusion_matrix(test_labels, predictions)
```

```python
#confusion matrix and summary report
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
print('Accuracy Score :',accuracy_score(test_labels, predictions))
print('Report : ')
print(classification_report(test_labels, predictions))
```
